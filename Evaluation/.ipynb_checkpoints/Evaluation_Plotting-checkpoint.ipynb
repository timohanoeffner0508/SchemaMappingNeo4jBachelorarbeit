```python
import numpy as np
import matplotlib.pyplot as plt

dfgrouping.plot.scatter(x='Durchgang', y='Zeit_in_Sekunden', color='yellow')

x = dfgrouping['Durchgang']
dsgrouping = dfgrouping['Zeit_in_Sekunden']

m, b = np.polyfit(x, dsgrouping, 1)
plt.plot(x, m * x + b, color='red', linestyle='-', label='Regression line')

mean_tiny = dsgrouping.mean()
std_tiny = dsgrouping.std()
median_tiny = dsgrouping.median()

plt.axhline(mean_tiny, color='blue', linestyle='--', label=f"Mean: {mean_tiny:.4f}")
plt.axhline(median_tiny, color='orange', linestyle='--', label=f"Median: {median_tiny:.4f}")
plt.axhline(mean_tiny + std_tiny, color='purple', linestyle=':', label=f"+1 Std: {mean_tiny + std_tiny:.4f}")
plt.axhline(mean_tiny - std_tiny, color='purple', linestyle=':', label=f"-1 Std: {mean_tiny - std_tiny:.4f}")

plt.legend()
plt.title("Benchmark-Auswertung (Tiny)")
plt.xlabel("Durchgang")
plt.ylabel("Zeit in Sekunden")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

dfmapping.plot.scatter(x='Durchgang', y='Zeit_in_Sekunden', color='yellow')

x1 = dfmapping['Durchgang']
dsmapping = dfmapping['Zeit_in_Sekunden']

m1, b1 = np.polyfit(x1, dsmapping, 1)
plt.plot(x1, m1 * x1 + b1, color='red', linestyle='-', label='Regression line')

mean_big = dsmapping.mean()
std_big = dsmapping.std()
median_big = dsmapping.median()

plt.axhline(mean_big, color='blue', linestyle='--', label=f"Mean: {mean_big:.4f}")
plt.axhline(median_big, color='orange', linestyle='--', label=f"Median: {median_big:.4f}")
plt.axhline(mean_big + std_big, color='purple', linestyle=':', label=f"+1 Std: {mean_big + std_big:.4f}")
plt.axhline(mean_big - std_big, color='purple', linestyle=':', label=f"-1 Std: {mean_big - std_big:.4f}")

plt.legend()
plt.title("Benchmark-Auswertung (Big)")
plt.xlabel("Durchgang")
plt.ylabel("Zeit in Sekunden")
plt.show()
```

```python
# Berechnung der Standardabweichung
std_main = pd.Series(dsmain_nw).std()
std_grouping = pd.Series(dsgrouping_nw).std()
std_mapping = pd.Series(dsmapping_nw).std()

plt.figure(figsize=(8, 5))
plt.plot(dsmain_nw, label='Export', marker='o', color='steelblue')
plt.axhline(y=pd.Series(dsmain_nw).mean(), color='steelblue', linestyle='--', linewidth=1)
plt.fill_between(range(len(dsmain_nw)),
                 pd.Series(dsmain_nw).mean() - std_main,
                 pd.Series(dsmain_nw).mean() + std_main,
                 color='steelblue', alpha=0.2
)

plt.plot(dsgrouping_nw, label='Grouping', marker='o', color='seagreen')
plt.axhline(y=pd.Series(dsgrouping_nw).mean(), color='seagreen', linestyle='--', linewidth=1)
plt.fill_between(range(len(dsgrouping_nw)),
                 pd.Series(dsgrouping_nw).mean() - std_grouping,
                 pd.Series(dsgrouping_nw).mean() + std_grouping,
                 color='seagreen', alpha=0.2)

plt.plot(dsmapping_nw, label='Mapping', marker='o', color='indianred')
plt.axhline(y=pd.Series(dsmapping_nw).mean(), color='indianred', linestyle='--', linewidth=1)
plt.fill_between(range(len(dsmapping_nw)),
                 pd.Series(dsmapping_nw).mean() - std_mapping,
                 pd.Series(dsmapping_nw).mean() + std_mapping,
                 color='indianred', alpha=0.2)

plt.xlabel('Durchlauf')
plt.ylabel('Zeit (s)')
plt.title('Laufzeitentwicklung mit Standardabweichung')
plt.legend(loc="best")


plt.show()
```

```python
# Daten vorbereiten
data = {
    "Datenbank": ["Dr Who", "IMDB", "Airbnb", "Stackoverflow", "Movie","Northwind"],
    "Knoten": [1060, 63042, 129444, 690656, 28863,1035],
    "Kanten": [2286, 106651, 220183, 1408205, 166261,3139],
    "E/V": [2.1566, 1.6917, 1.70099, 2.0389, 5.7603,3.0328],
    "Zeit (s)": [1.064, 53.084, 108.935, 556.656, 43.3211, 0.103444]
}

df = pd.DataFrame(data)

# Visualisierung: Balkendiagramm der Extraktionszeit und Linienplot des E/V-Verhältnisses
fig, ax1 = plt.subplots(figsize=(10, 6))

# Balken für Extraktionszeit
bar = ax1.bar(df["Datenbank"], df["Zeit (s)"], color="seagreen", label="Extraktionszeit (s)")
ax1.set_ylabel("Extraktionszeit (s)")
ax1.set_xlabel("Datenbank")
ax1.set_title("Vergleich der Extraktionszeiten und Beziehungsdichte")
ax1.tick_params(axis='y')

# Linienplot für E/V-Verhältnis
ax2 = ax1.twinx()
line = ax2.plot(df["Datenbank"], df["E/V"], color="red", marker="o", label="E/V-Verhältnis")
ax2.set_ylabel("E/V-Verhältnis")

# Legenden kombinieren
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc="upper left")

plt.tight_layout()
plt.show()
```

```python

data = {
    "Datenbank": ["Dr Who", "IMDB", "Airbnb", "Stackoverflow", "Movie", "Northwind"],
    "Knoten": [1060, 63042, 129444, 690656, 28863,1035],
    "Kanten": [2286, 106651, 220183, 1408205, 166261,3139],
    "Laufzeit (s)": [1.064, 53.084, 108.935, 556.656, 43.3211,0.103444]
}

df = pd.DataFrame(data)


df["Extraktionsrate (Objekte/s)"] = (df["Knoten"] + df["Kanten"]) / df["Laufzeit (s)"]


plt.figure(figsize=(6, 6))
bars = plt.bar(df["Datenbank"], df["Extraktionsrate (Objekte/s)"], color="slateblue", width=0.2)
plt.xlabel("Datenbank")
plt.ylabel("Extraktionsrate (Objekte pro Sekunde)")


for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 20, f"{yval:.0f}", ha='center', va='bottom', fontsize=9)

plt.tight_layout()
```

```python
# Berechnung der Standardabweichung
std_main_nw = pd.Series(dsmain_nw).std()
std_grouping_nw = pd.Series(dsgrouping_nw).std()
std_mapping_nw = pd.Series(dsmapping_nw).std()

std_main = pd.Series(dsmain).std()
std_grouping = pd.Series(dsgrouping).std()
std_mapping = pd.Series(dsmapping).std()


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 7), sharey=False)

# Erster Plot: Normalisiert
ax1.plot(dsmain_nw, label='Export', marker='o', color='steelblue')
ax1.axhline(y=pd.Series(dsmain_nw).mean(), color='steelblue', linestyle='--', linewidth=1)
ax1.fill_between(range(len(dsmain_nw)),
                 pd.Series(dsmain_nw).mean() - std_main_nw,
                 pd.Series(dsmain_nw).mean() + std_main_nw,
                 color='steelblue', alpha=0.2)

ax1.plot(dsgrouping_nw, label='Grouping', marker='o', color='seagreen')
ax1.axhline(y=pd.Series(dsgrouping_nw).mean(), color='seagreen', linestyle='--', linewidth=1)
ax1.fill_between(range(len(dsgrouping_nw)),
                 pd.Series(dsgrouping_nw).mean() - std_grouping_nw,
                 pd.Series(dsgrouping_nw).mean() + std_grouping_nw,
                 color='seagreen', alpha=0.2)

ax1.plot(dsmapping_nw, label='Mapping', marker='o', color='indianred')
ax1.axhline(y=pd.Series(dsmapping_nw).mean(), color='indianred', linestyle='--', linewidth=1)
ax1.fill_between(range(len(dsmapping_nw)),
                 pd.Series(dsmapping_nw).mean() - std_mapping_nw,
                 pd.Series(dsmapping_nw).mean() + std_mapping_nw,
                 color='indianred', alpha=0.2)

ax1.set_title("Transformationsprozess Northwind-Datenbank")
ax1.set_xlabel("Durchlauf")
ax1.set_ylabel("Zeit (s)")

# Zweiter Plot: Originaldaten
ax2.plot(dsmain, marker='o', color='steelblue')
ax2.axhline(y=pd.Series(dsmain).mean(), color='steelblue', linestyle='--', linewidth=1)
ax2.fill_between(range(len(dsmain)),
                 pd.Series(dsmain).mean() - std_main,
                 pd.Series(dsmain).mean() + std_main,
                 color='steelblue', alpha=0.2)

ax2.plot(dsgrouping, marker='o', color='seagreen')
ax2.axhline(y=pd.Series(dsgrouping).mean(), color='seagreen', linestyle='--', linewidth=1)
ax2.fill_between(range(len(dsgrouping)),
                 pd.Series(dsgrouping).mean() - std_grouping,
                 pd.Series(dsgrouping).mean() + std_grouping,
                 color='seagreen', alpha=0.2)

ax2.plot(dsmapping, marker='o', color='indianred')
ax2.axhline(y=pd.Series(dsmapping).mean(), color='indianred', linestyle='--', linewidth=1)
ax2.fill_between(range(len(dsmapping)),
                 pd.Series(dsmapping).mean() - std_mapping,
                 pd.Series(dsmapping).mean() + std_mapping,
                 color='indianred', alpha=0.2)

ax2.set_title("Transformationsprozess Movie-Datenbank")
ax2.set_xlabel("Durchlauf")

# Gemeinsame Legende
handles, labels = ax1.get_legend_handles_labels()
fig.legend(handles, labels, loc='upper center', ncol=3)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
